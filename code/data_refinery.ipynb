```python
import pandas as pd
import numpy as np
from pathlib import Path
import os
```

```python
files = []
folder_path = Path(os.getcwd()+"/../csvFiles")
for file in folder_path.glob('*.csv'):
    files.append(file)
for x in range(len(files)):
    print(str(x)+": "+str(files[x]))
```

```python
merged_data = pd.read_csv(files[3])
```

```python
merged_data
```

```python
ls = []
for col in merged_data:
    null_percentage = merged_data[col].isnull().mean() * 100
    print(null_percentage)
    if null_percentage > 60:
        ls.append(col)
for col in ls:
    merged_data = merged_data.drop(col,axis = 1)
```

```python
merged_data.to_csv("r_merged_data.csv")
```
