```python
import pandas as pd
from pathlib import Path
import json
import os
import numpy as np
```

```python
def extract_data(data):
    out = []
    d = data['data']['results']
    for item in d:
        ls = []
        l = item['location']
        try:
            ls.append(l['address']['state'])
        except TypeError as e:
            ls.append(np.nan)
        try:
            ls.append(l['address']['coordinate']['lat'])
        except TypeError as e:
            ls.append(np.nan)
        try:
            ls.append(l['address']['coordinate']['lon'])
        except TypeError as e:
            ls.append(np.nan)
        try:
            ls.append(l['county']['name'])
        except TypeError as e:
            ls.append(np.nan)
        out.append(ls)
    print(len(out))
    return out
```

```python

folder_path = Path(os.getcwd()+"/data")
dataList = []
for file in folder_path.glob("*.json"):  # Loops through all JSON files
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        s = extract_data(data)
        dataList.append(s)
```

```python
df = pd.DataFrame(columns=['city','county','lat','lon'])
for list in dataList:
    for l in list:
        a = l[0]
        b = l[3]
        c = l[1]
        d = l[2]
        ls = [a,b,c,d]
        df.loc[len(df)] = ls
df.head()
```
